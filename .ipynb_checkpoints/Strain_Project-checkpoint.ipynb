### Training with DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
```

### We remove column Metallic Materials from our data beacuse it is a categorical string feature

```python
X = data.iloc[:, 0:-1].values
y = data.iloc[:, -1].values
```

```python
X[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17) 
```

```python
tree.fit(X=X_train, y=y_train)
```

```python
y_predicted = tree.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_predicted)
```

```python
np.sqrt(mean_squared_error(y_test, y_predicted))
```

```python
df = pd.DataFrame({
    'y_predicted_elastic_modulus' : y_predicted,
    'y_test_elastic_modulus' : y_test
})
```

```python
df
```

### To remove a column from data, and then check it's mean_squared_error

```python
column_name = data.columns
```

```python
dct = {
    'column_removed' : [],
    'rmse' : []
}
```

```python
df = pd.DataFrame(dct)
```

```python
for i in range(0, X.shape[1]):
    # Removing the ith column, how much mse do we get?
    
    Xnew_scaled = np.delete(X_scaled, i, 1)
    tree = DecisionTreeRegressor()
    X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(Xnew_scaled, y, test_size=0.3, random_state=17)
    tree.fit(X=X_train_scaled, y=y_train)
    y_predicted = tree.predict(X_test_scaled)
    df.loc[df.shape[0]] = [column_name[i], np.sqrt(mean_squared_error(y_predicted, y_test))]
```

```python
df
```

```python
df.sort_values('rmse')
```
